## Feature Engineering for E-commerce Customer Behavior Analysis
The goal of this notebook is to preprocess the dataset and create meaningful features that capture user behavior and purchase intent for further KPI analysis and dashboard development.

```python
import pandas as pd
```

```python
# read the dataset
data = pd.read_csv('online_shoppers_intention.csv')
data.shape
```

```python
# check the data type and missing values or duplicates
data.info()
data.duplicated().sum()
data.isna().sum()
data.head(5)
```

### Feature Engineering
In this section, we create features to better capture user engagement, browsing behavior, and purchase intent.

```python
# Feature Engineering
# 1.1 calculate total time spening on the web
data['total_duration'] = (data['Administrative_Duration'] + data['Informational_Duration'] + data['ProductRelated_Duration']).round(2)

# 1.2 calculate the total number of pages
data['total_pages'] = data['Administrative'] + data['Informational'] + data['ProductRelated']

# 1.3 calculate the average time spend on one page
data['avg_duration_per_page'] = (data['total_duration'] / (data['total_pages'].replace(0,1))).round(2)
```

```python
# 2. calculate product engagement rate
data['product_page_ratio'] = data['ProductRelated'] / (data['total_pages'].replace(0,1))

# 3. categorize visitors based on the duration spend on product related pages
data['productrelated_intent'] = pd.qcut(data['ProductRelated_Duration'], q=4, labels=[
    'low', 'medium', 'high', 'very_high'], duplicates='drop')
data['total_duration_level'] = pd.qcut(data['total_duration'], q=4, labels=[
    'low', 'medium', 'high', 'very_high'], duplicates='drop')
```

```python
# 4. convert binary columns to 0/1
data['Weekend'] = data['Weekend'].astype(int)
data['Revenue'] = data['Revenue'].astype(int)
data['is_returning'] = (data['VisitorType']=='Returning_Visitor').astype(int)

# 5. convert month column from text to int
month_map = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
    'May': 5, 'June': 6, 'Jul': 7, 'Aug': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
data['Month'] = data['Month'].map(month_map)
```

## Summary
I created several features to capture user engagement and intent, including:
- Total browsing duration
- Average duration per page
- Product page ratio
- Intent segmentation

The processed dataset is now ready for SQL-based KPI analysis and dashboard development.

```python
data.to_csv('processed_online_shoppers_intention.csv')
```
